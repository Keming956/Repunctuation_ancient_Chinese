```python
import os
import re
import string

folder_path = input("Entrez le chemin du dossier : ")

# Créer le dossier de sortie
output_folder = os.path.join(folder_path, "Traités")
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Parcourir tous les fichiers 
processed_files = 0
for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        file_path = os.path.join(folder_path, filename)
        output_path = os.path.join(output_folder, filename)
        
        try:
            with open(file_path, 'r', encoding='utf-8') as file:
                lines = file.readlines()
            
            # Trouver l'index de la dernière ligne contenant un *
            last_star_line_index = -1
            for i, line in enumerate(lines):
                if '*' in line:
                    last_star_line_index = i
            
            # Si une ligne avec astérisque est trouvée, ne garder que le contenu après cette ligne
            if last_star_line_index != -1:
                filtered_lines = lines[last_star_line_index + 1:]
            else:
                # Si aucune ligne avec astérisque n'est trouvée, garder tout le contenu
                filtered_lines = lines
                print(f"Attention : Aucune ligne contenant '*' n'a été trouvée dans {filename}")
            
            # Filtrer les lignes contenant les flèches ← ou →
            filtered_lines = [line for line in filtered_lines if '←' not in line and '→' not in line]
            
            # Joindre toutes les lignes et supprimer les sauts de ligne et les espaces triples
            content = ''.join(filtered_lines).replace('\n', '').replace('   ', '')
            
            # Écrire le contenu traité dans un nouveau fichier
            with open(output_path, 'w', encoding='utf-8') as file:
                file.write(content)
            
            processed_files += 1
            print(f"Traité: {filename}")
        
        except Exception as e:
            print(f"Erreur lors du traitement de {filename}: {str(e)}")

print(f"\nTraitement terminé! {processed_files} fichiers ont été traités.")
print(f"Les fichiers traités sont enregistrés dans: {output_folder}")
```

```python

# Définir l'ensemble des signes de ponctuation en chinois traditionnel
chinese_punctuation = "，。、！？：；「」『』''""（）【】《》〈〉…—～·*   "
# Combiner les signes de ponctuation chinois et anglais
all_punctuation = string.punctuation + chinese_punctuation

# Spécifier le chemin 
folder_path = input("Entrez le chemin du dossier : ")

# Créer le dossier de sortie
output_folder = os.path.join(folder_path, "Sans_Ponctuation")
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Parcourir tous les fichiers txt dans le dossier
processed_files = 0
for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        file_path = os.path.join(folder_path, filename)
        output_path = os.path.join(output_folder, filename)
        
        try:
            # Lire le contenu du fichier
            with open(file_path, 'r', encoding='utf-8') as file:
                content = file.read()
            
            # Supprimer tous les signes de ponctuation
            for char in all_punctuation:
                content = content.replace(char, '')
            
            # Écrire le contenu traité dans un nouveau fichier
            with open(output_path, 'w', encoding='utf-8') as file:
                file.write(content)
            
            processed_files += 1
            print(f"Traité : {filename}")
        
        except Exception as e:
            print(f"Erreur lors du traitement de {filename} : {str(e)}")

print(f"\nTraitement terminé ! {processed_files} fichiers ont été traités.")
print(f"Les fichiers traités sont enregistrés dans : {output_folder}")
```
